Project Goal: Be able to predict the amount of time for a car to sell. Develop an understanding of what features make up a car that sells fast (35 days or less).

What's been done so far:

Data Munging: Removed columns with null features and replaced values for columns that had some nulls. Created additional features such as age group, primary car class, and secondary car class.

Analysis: Used visualizations to determine how to split age group and to take an overall look at the makeup of each age group.

Modeling: Ran lasso to see the mean square error when it comes to predicting age and random forest for age group. Also wanted to try KNN to find the charateristics of age_group = fast and maybe KMeans clustering to see interesting clusters.

```python
import pandas as pd
import numpy as np
import math
from sklearn.ensemble.forest import RandomForestRegressor
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from patsy import dmatrices, dmatrix
from sklearn.linear_model import Lasso
from sklearn.cross_validation import KFold, cross_val_score
from sklearn.dummy import DummyClassifier, DummyRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import RandomizedPCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('merch_data.csv')
#pd.set_option('display.max_rows', 500)
```

```python
df.columns
```

```python
# clean up column names
df.columns = df.columns.str.replace(' ','_')
df.columns = [x.lower() for x in df.columns]
```

```python
df.info()
```

```python
# remove feature if null count is above 1000
null_cnts = df.isnull().sum()
headers = []
for x in null_cnts.keys():
    if null_cnts[x]<=1000:
        headers.append(x)

clean = df[headers]
```

```python
# fill in null values 
clean['_free_shipping_cost_'] = clean['_free_shipping_cost_'].fillna(0)
clean['trade_acv'] = clean['trade_acv'].fillna(0)
clean = clean.fillna('missing, missing')
clean.head()
```

```python
# reduce number of sales sources by grouping into broader groups
def update_sales_source(x):
    if x in('repeat customer','referral','dealer','known other','truecar','craigslist'):
        return 'known_other'
    elif x in('unknown','website','drive by'):
        return 'unknown'
    else:
        return x

clean['sales_source'] = clean.sales_source.apply(lambda x: x.lower())
clean['sales_source'] = clean.sales_source.apply(update_sales_source)
```

```python
# create additional features from car class
clean['car_class'].value_counts()
```

```python
clean['primary_class'] = clean.car_class.apply(lambda x: x.split(', ')[0].split()[0])
clean['secondary_class'] = clean.car_class.apply(lambda x: x.split(', ')[1].split()[0])
```

```python
# create acquisition source feature (where we got the car)
def acq_source(x):
    if str(x)[0] in('1','2'):
        return 'consumer_trade'
    elif str(x)[0] in('3','4','5'):
        return 'auction_purchase'
    elif str(x)[0] == '7':
        return 'dealer_purchase'
    elif str(x)[0] in('8','9'):
        return 'consumer_purchase'
    
clean['acq_source'] = clean.stock_number.apply(acq_source)
```

```python
clean.info()
```

```python
# see how is age of sale distributed
clean.age.hist(bins=25)
```

```python
# based on visualization, group ages into buckets
def age_groups(x):
    if x <= 35:
        return 'fast'
    elif x <=60:
        return 'good'
    elif x <=90:
        return 'average'
    else:
        return 'poor'

clean['age_group'] = clean.age.apply(age_groups)
```

```python
# see how distance from dealership is distributed
clean.distance.hist(bins=30)
```

```python
# based on visualization, distance from dealership into buckets
def dealer_range(x):
    if x <= 100:
        return '0-100'
    else:
        return '100+'
    
clean['dealer_range'] = clean.distance.apply(dealer_range)
```

```python
# get features that are ints and floats
num_features = []
for x in clean.columns:
    if clean[x].dtype in ('int64','float64'):
        num_features.append(x)
```

```python
num_features = ' + '.join(num_features)
num_features
```

```python
# remove features that we know AFTER the car is sold i.e. gross profit etc...
# this shouldn't help us determine how fast a car sells
tweaked_num = 'distance + _free_shipping_cost_ + year + mileage + trade_acv + acquisition_cost + reconditioning_cost + freight + ninetyday + total_cost + sale_price + financing + service_contract + gap + permaplate'
```

```python
# get features that are strings
obj_features = []
for x in clean.columns:
    if clean[x].dtype == 'object':
        obj_features.append(x)
```

```python
# remove age group since we're also trying to predict this
obj_features.remove('age_group')
```

```python
# make these features categorical
obj_features = ') + C('.join(obj_features)
```

```python
obj_features
```

```python
# remove features like stockid, zipcode and buyer state (already measured by distance)
tweaked_obj = '+ C(sales_source) + C(car_class) + C(body) + C(color) + C(trim) + C(make) + C(model) + C(primary_class) + C(secondary_class) + C(acq_source) + C(dealer_range)'
```

```python
# does dmatrices do the samething as pd.factorize(data['x']) to categoricals????
y, X = dmatrices('age ~ 0 +' + tweaked_num + tweaked_obj , clean, return_type = 'dataframe')
#flatter y to a list
y = np.ravel(y)
```

```python
# fit data to the model first try
ls = Lasso(max_iter=100)
ls.fit(X, y) #<- predicting y given X
np.abs(cross_val_score(ls, X, y, scoring='mean_absolute_error', cv=10).mean())
```

```python
# scale data to improve lasso cv score
scaler = StandardScaler(copy=True)
X_scaled = scaler.fit(X).transform(X)
y_scaled = scaler.fit(y).transform(y)
```

```python
# fit data to the model 2nd try
# result is 0.8 sd off from mean
ls = Lasso(max_iter=100) #scales features
ls.fit(X_scaled, y_scaled) #<- predicting y given X
np.abs(cross_val_score(ls, X_scaled, y_scaled, scoring='mean_absolute_error', cv=10).mean())
```

```python
data = dmatrix(formula + ' + 0', clean, return_type='dataframe')
```

```python
data.head()
```

```python
scale = StandardScaler()
scaled_data = scale.fit(data).transform(data)
X_scale = scale.fit(data).transform(data)
```

```python
km = KMeans(n_clusters=3)
km.fit(scaled_data)
```

```python
print km.inertia_
```

```python
x = []
y = []
for n in range(1,20):
    km2 = KMeans(n_clusters=n)
    km2.fit(X_scale)
    x.append(n)
    y.append(km2.inertia_)
plt.plot(x, y)
```

```python
# rerun with 7 based on elbow around 7
km = KMeans(n_clusters=7)
km.fit(scaled_data)
```
